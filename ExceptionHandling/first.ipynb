Exception Handling

```python
try:
    a=10
    b=0
    c=a/b
except Exception as e:#Generic Exception Handling
    
    print("Exception message:", e)
```

```python
try:
    a=10
    b=0
    c=a/b
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

```python
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b
except ZeroDivisionError as e:
    print(e)
```

Type Error

```python
try:
    a = 10
    b = "20"
    print(a + b)
except TypeError:
    print("Type Error! Cannot add int and string.")
```

Value Error

```python
try:
    num = int(input("Enter a number: "))
    print(num)
except ValueError:
    print("Invalid value! Please enter a number.")
```

```python
#Indexed Error
try:
    list1 = [1, 2, 3]
    print(list1[5])
except IndexError:
    print("Index Error! List index out of range.")
```

```python
#KeyError
try:
    dict1 = {"name": "Alice", "age": 30}
    print(dict1["gender"])
except KeyError:
    print("Key Error! Key not found in the dictionary.")
```

```python
#NameError
try:    
    print(x)
except NameError:
    print("Name Error! Variable 'x' is not defined.")
```

```python
#FileNotFoundError
try:
    x=open("non_existent_file.txt")
except FileNotFoundError:
    print("File Not Found Error! The specified file does not exist.")
```

```python
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b
except ZeroDivisionError as e:
    print(e)
except ValueError as e:
    print(e)
```

```python
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b
    print("Result:", c)
except (ZeroDivisionError, ValueError) as e:
    print(e)
```

```python
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b
except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed.")
    print("Exception message:", e)
else:
    print("Program executed successfully. Result:")
```

```python
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b
except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed.")
    print("Exception message:", e)
else:
    print("Program executed successfully. Result:")
finally:
    print("Program Finished")
```
