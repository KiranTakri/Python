class emp
emp_name - public
emp_dept - protected
emp_salary-private
using get access the salary
using set update the salary 

```python
class emp:
    def __init__(self,emp_name,emp_dept,emp_salary):
        self.emp_name=emp_name
        self._emp_dept=emp_dept
        self.__emp_salary=emp_salary
    def get_emp_salary(self):
        return self.__emp_salary
    def set_emp_salary(self,emp_salary):
        self.__emp_salary=emp_salary
class empInfo(emp):
    def display(self):
        print("Employee Name:",self.emp_name)
        print("Employee Department:",self._emp_dept)
        print("Employee Salary:",self.get_emp_salary())
emp1=empInfo("Kiran","IT",50000)
emp1.display()
emp1.get_emp_salary()
emp1.set_emp_salary(60000)
emp1.display()
```

```python
class emp:
    def __init__(self,emp_name,emp_dept,emp_salary):
        self.emp_name=emp_name
        self._emp_dept=emp_dept
        self.__emp_salary=emp_salary
    def get_emp_salary(self):
        return self.__emp_salary
    def set_emp_salary(self,emp_salary):
        self.__emp_salary=emp_salary
class manager(emp):
    
    def show(self):
        print("Name:",self.emp_name)
        print("Department:",self._emp_dept)
        print("Salary:",self.get_emp_salary())
    def update_salary(self,emp_salary):
        self.set_emp_salary(emp_salary)
manager1=manager("Kiran","IT",50000)
manager1.show()
manager1.update_salary(60000)
manager1.show()
```
