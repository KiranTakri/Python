```python
class number:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        x=self.num
        if x<5:
         self.num+=1
         return x
        else:
           raise StopIteration
n=number()
it=iter(n)

while True:
   try:
    print(next(it))
   except StopIteration as s:
      print(s)
      break
   

    


```

Generator- It is a subclass of Iterator


```python
def number():
    yield 1
    yield 2
    yield 3
n=number()
it=iter(n)
print(next(it))
print(next(it))
print(next(it))
```

```python
def number():
    yield 1
    yield 2
    yield 3
n=number()
for i in n:
    print(i)
```

```python
def number():
    yield 1
    yield 2
    yield 3
for i in number():
    print(i)
```

```python
def number():
    for i in range(1,11):
        yield i
for j in number():
    print(j)
```

```python
def number():
    for i in range(1,11):
        yield i*i
for j in number():
    print(j)
```

```python
l=[10,20,30,50]
print(any(i>=40 for i in l))
print(all(i>=10 for i in l))
```
